```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic= sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.set_style('darkgrid')
sns.jointplot(y='age', x='fare', data=titanic, kind='reg')
```

```python

sns.distplot(titanic['survived'], kde=True)
```

```python
sns.distplot(titanic['fare'], kde=False, color='red')
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x='class', y='age', data=titanic)
```

```python
#titanic.head(2)
sns.barplot(x='sex', y='fare', data=titanic)
```

```python
count= titanic.groupby('survived').count()
count
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=titanic)
```

```python
tit_cor=titanic.corr()
```

```python
sns.heatmap(tit_cor, cmap='coolwarm', lw=0.5)
plt.title('titanic.corr()')
```

```python
sns.set_style('whitegrid')
g =sns.FacetGrid(data=titanic, col='sex')
g.map(sns.distplot, 'age', kde=False, color='blue')
```
