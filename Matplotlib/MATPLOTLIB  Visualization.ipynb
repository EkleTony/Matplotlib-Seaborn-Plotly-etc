```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
#FUNCTIONAL METHOD
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
#plt.show() # functions like a print func
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2) # 1 row, by two column and 2 plot num
plt.plot(x,y, 'b')
```

# object oriented method

```python
fig = plt.figure()

axes =  fig.add_axes([0.1,0.1, 0.8, 0.8]) #left, bottom,width and height ofthe axes
axes.plot(x,y, 'r')
axes.set_xlabel('X label')
axes.set_ylabel( 'Y label')
axes.set_title('Set title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.6,0.3,0.2])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('Smaller plot')
axes1.set_title('Larger plot')
```

# Matplotlib part 2

```python
#SUBPLOT
```

```python
fig, axes = plt.subplots()

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second PLot')
```

```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    #ax.set_title('Title')
    
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y , 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

fig
plt.tight_layout()
```

# figure size, aspect ratio and DPI

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(x,y)
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X squared')
ax.set_xlabel('x')
ax.set_title('Title')
ax.set_ylabel('y')


ax.plot(x,x**3, label='Y Cubed')

ax.legend(loc=0) # zero is best location
#ax.legend(loc=(0.1,0.4))
```

# PLOT APPEARANCE

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', lw=2, ls=':')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
